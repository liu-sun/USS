```python
import json
import pprint

from Bio import Entrez

Entrez.email = ''
Entrez.api_key = ""
term = '"11586351"[PubMed ID]'
handle = Entrez.esearch(db='clinvar', term=term, usehistory='y')
record = Entrez.read(handle)
webenv = record['WebEnv']
query_key = record['QueryKey']
handle = Entrez.esummary(db='clinvar', query_key=query_key, retmode='json', webenv=webenv)
pprint.pprint(json.loads(handle.read().decode('utf8')))
```
